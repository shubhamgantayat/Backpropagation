```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

```python
class Activation:
    
    @staticmethod
    def sigmoid(val, l=1):
        return 1 / (1 + np.exp(-1*l*val))

    
class NN:
    
    def __init__(self, X, y, hidden_neurons=64, learning_rate=0.001):
        self.X = X
        self.y = y
        self.batch_size = len(self.X)
        self.input_neurons = self.X.shape[1]
        self.output_neurons = self.y.shape[1] if len(self.y.shape) == 2 else 1
        self.hidden_neurons = hidden_neurons
        self.learning_rate = learning_rate
        self.weights1 = np.random.random((self.input_neurons, self.hidden_neurons)) * 1e-8
        self.weights2 = np.random.random((self.hidden_neurons, self.output_neurons)) * 1e-8
        self.weights1_grad = np.ndarray((self.input_neurons, self.hidden_neurons, self.batch_size))
        self.weights2_grad = np.ndarray((self.hidden_neurons, self.output_neurons, self.batch_size))
        self.activation = np.vectorize(Activation().sigmoid)
        
    def forward_pass(self):
        self.hidden_input = self.X@self.weights1
        self.hidden_output = self.activation(self.hidden_input)
        self.output_input = self.hidden_output@self.weights2
        self.output_output = self.activation(self.output_input)
    
    def calculate_error(self):
        self.error = mean_squared_error(self.y, self.output_output)
    
    def backward_pass(self):
        for i in range(self.output_neurons):
            for j in range(self.hidden_neurons):
                self.weights2_grad[j][i] = -1 * (self.y[:, i] - self.output_output[:, i]) * self.output_output[:, i] * (1 - self.output_output[:, i]) * self.hidden_output[:, j]
        tmp = np.sum(nn.weights2_grad, axis=1).T
        for j in range(self.hidden_neurons):
            for k in range(self.input_neurons):
                self.weights1_grad[k][j] = tmp[:, j] * self.hidden_output[:, j] * (1 - self.hidden_output[:, j]) * self.X[:, k]
    
    def update_weights(self):
        self.weights1 = self.weights1 - self.learning_rate * np.mean(self.weights1_grad, axis=2)
        self.weights2 = self.weights2 - self.learning_rate * np.mean(self.weights2_grad, axis=2)
```

```python
data = load_breast_cancer()
X = data.data
y = np.c_[data.target]
```

```python
nn = NN(X, y, learning_rate=1e-4)
epochs = 10000
for epoch in range(epochs):
    print(f"Epoch : {epoch+1} / {epochs}")
    nn.forward_pass()
    nn.calculate_error()
    nn.backward_pass()
    nn.update_weights()
    print(nn.error)
    print("______________________________________")
```
